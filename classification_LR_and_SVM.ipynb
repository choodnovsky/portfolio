```python
import numpy as np
import matplotlib.pyplot as plt
from pylab import *
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

```python
X = np.random.rand(100, 1)
y = 4 + 5 * X + 1.0 * np.random.rand(100, 1)

reg = LinearRegression()
reg.fit(X, y)

X_vals = np.linspace(0, 1, 100).reshape(-1, 1)
y_vals = reg.predict(X_vals)

plt.scatter(X, y)
plt.plot(X_vals, y_vals, color='r')
plt.show()
```

```python
X = 4 * np.random.rand(100, 1) - 2
y = 4 + 2 * X + 5 * X**2 + 30. * np.random.rand(100, 1)

poly_features = PolynomialFeatures(degree=5, include_bias=False)
X_poly = poly_features.fit_transform(X)

reg = LinearRegression()
reg.fit(X_poly, y)

X_vals = np.linspace(-2, 2, 100).reshape(-1, 1)
X_poly_vals = poly_features.fit_transform(X_vals)
y_vals = reg.predict(X_poly_vals)

plt.scatter(X, y)
plt.plot(X_vals, y_vals, color='r')
plt.show()
```

#### Пример 2  
_Количество плохих отзывов обратно зависит от ср количества покупок_

```python
np.random.seed(2)
# Покупки
purchases = np.random.normal(3.0, 1.0, 1000) # среднее=3, дисперсия=1, значений=1000
# плохие оценки
negatives = np.random.normal(10.0, 1.0, 1000) / purchases
```

```python
scatter(purchases, negatives)
plt.show()
```

_Переменные для обучения_

```python
x = np.array(purchases)
y = np.array(negatives)
```

_Полином 2 степени или парабола_

```python
p2 = np.poly1d(np.polyfit(x, y, 2)) # задаем полином 2 степени подогнанный под два массива
xp = np.linspace(0, 7, 100) # выводим как линию из ста точек
plt.scatter(x, y) # выводим скаттерплот двух массивов
plt.plot(xp, p2(xp), c='r') # 
plt.show()
```

```python
r2 = r2_score(y, p2(x))
r2
```

_Полином 5 степени_

```python
p5 = np.poly1d(np.polyfit(x, y, 5)) # задаем полином 2 степени подогнанный под два массива
xp = np.linspace(0, 7, 100) # выводим как линию из ста точек
plt.scatter(x, y) # выводим скаттерплот двух массивов
plt.plot(xp, p5(xp), c='r') # 
plt.show()
```

```python
r2 = r2_score(y, p5(x))
r2
```

```python
import numpy as np
import pandas as pd
```

```python
orders = pd.read_excel('/Users/victorchudnovskiy/Documents/orders.xlsx', sheet_name='Заказы')
new_order = pd.read_excel('/Users/victorchudnovskiy/Documents/orders.xlsx', sheet_name='Новый', na_values='#Н/Д')
```

```python
def recomend(df, new_order):
    data = []
    new_order['cnt'] = new_order['cnt'].fillna(0)
    new = new_order.set_index('Article')
    df_prep = df.pivot_table(
        index='Order_Number', 
        columns='Article', 
        values='cnt', 
        aggfunc='sum', fill_value=0).T.join(new).T
    b = df_prep.iloc[-1:].to_numpy()[0]
    bLength = np.linalg.norm(b)
    for i in df_prep.index:
        a = df_prep.loc[i:].to_numpy()[0]
        aLength = np.linalg.norm(a)
        res = np.dot(a, b) / (aLength * bLength)
        data.append((i, res))
    data = pd.DataFrame(data, columns=['order', 'metrica'])
    orders = data.query('1 > metrica > 0.6')['order']
    data = (df_prep.loc[orders].max() - new['cnt']).to_frame()
    data.columns = ['recomend']
    return pd.concat([new[new['cnt'] !=0], data[data['recomend'] > 0.0]])
```

```python
df = recomend(orders, new_order)
```
